```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import requests
import numpy as np
import pandas as pd

r = requests.get('http://coronosis.live/api/corona/overlapping_curves/BD/affected')

current =[]
one_day = []
two_day = []

predictions = r.json()

for i in range(len(predictions)):
    current.append(predictions[i]['current'])
    one_day.append(predictions[i]['one_day'])
    two_day.append(predictions[i]['two_day'])

print(current[0],one_day[0],two_day[0])

array1 = np.array(current)
array2 = np.array(one_day)
array3 = np.array(two_day)

difference_arrayone = np.subtract(array1, array2)
absoluteone = np.abs(difference_arrayone)
print("mean absolute error for one day affetced")
maeone = np.mean(absoluteone)
print(maeone)
squarediffone = np.square(difference_arrayone)
print("root mean square error for one day affetced")
mseone = np.mean(squarediffone)
rmseone = np.sqrt(mseone)
print(rmseone)
percentage_arrayone = np.true_divide(difference_arrayone, array1)
squared_arrayone = np.square(percentage_arrayone)
print("mean squared percentage error for one day affetced")
mspeone = np.mean(squared_arrayone) *100
print(mspeone)

print("root mean squared percentage error for one day affetced")
rmspeone = np.sqrt(np.mean(squared_arrayone)) *100
print(rmspeone)
# print(np.sqrt(mean_squared_pererror(current, one_day)))
# print(np.sqrt(mean_squared_error(current, two_day)))
# print('here')

# print(mean_absolute_error(current, one_day))
# print(mean_absolute_error(current, two_day))
print("Latest Absolute percentage error for one day prediction affected")
print((np.abs(current[len(current)-1] - one_day[len(one_day)-1])/current[len(current)-1])*100)
      
difference_arraytwo = np.subtract(array1, array3)
percentage_arraytwo = np.true_divide(difference_arraytwo, array1)
squared_arraytwo = np.square(percentage_arraytwo)
print("mean squared percentage error for two day affetced")
mspetwo = np.mean(squared_arraytwo) *100
print(mspetwo)

print("root mean squared percentage error for two day affetced")
rmspetwo = np.sqrt(np.mean(squared_arraytwo)) *100
print(rmspetwo)
# print(np.sqrt(mean_squared_pererror(current, one_day)))
# print(np.sqrt(mean_squared_error(current, two_day)))
# print('here')

# print(mean_absolute_error(current, one_day))
# print(mean_absolute_error(current, two_day))
print("Latest Absolute percentage error for two day prediction affected")
print((np.abs(current[len(current)-1] - two_day[len(two_day)-1])/current[len(current)-1])*100)
```

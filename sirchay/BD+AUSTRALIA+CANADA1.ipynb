```python
df2 = pd.read_csv('BD1.csv')
df2
```

```python
df2 = df2.set_index(pd.to_datetime(df2.Date), drop=True)
df2.ActualAffected.plot(grid=True, label="Actual", legend=True)
df2.One_day_prediction_of_affected.plot(grid=True, label="One day", legend=True)
df2.Two_day_prediction_of_affected.plot(grid=True, label="Two day", legend=True)
plt.legend(loc='best')
plt.title('Actual and Predicted Affected Cases of Bangladesh for One and Two Days Prediction')
fig.savefig('BDaffected1and2.png',dpi=300, bbox_inches='tight')
plt.show()
```

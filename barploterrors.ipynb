```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib.ticker import MultipleLocator
 
# barWidth = 0.5
# Read CSV into pandas 
majorloc = MultipleLocator(1)
df = pd.read_csv("Error_eliaffected.csv") 
df = df.sort_values(by=['country_code'])
name = df.iloc[0:69,0:1] 
# error = df['rootmeansquarep1']
error = df.iloc[0:69,1:2]
n = [x[0] for x in name.values.tolist()]
e = [y[0] for y in error.values.tolist()]
name = []
for i in range(0,len(e),2):
    name.append(n[i])
r1 = np.arange(len(e))
# Figure Size 
fig = plt.figure(figsize =(100, 40)) 
fig = plt.gcf()
r = 0
# Horizontal Bar Plot 
plt.bar(r1,e, color='#597E85', width=.4, edgecolor='white')
plt.grid(b=True, which='major', color ='black', axis='both', linestyle=':', linewidth=1.5)
 
# Add xticks on the middle of the group bars
plt.xlabel('Country code',fontname='Times new roman',fontsize='130')
plt.xticks(np.arange(0, 57, step=2),name,fontname='Times new roman',fontsize='96', rotation = 45)
plt.yticks(fontname='Times new roman',fontsize='96')
plt.ylabel('Average percentage prediction error',fontname='Times new roman',fontsize='130')
 
# Create legend & Show graphic
# plt.legend()

plt.show()
plt.draw()
fig.savefig('erroraffone.pdf', dpi=600, bbox_inches='tight')
  
# # Show Plot 
# plt.show() 
```
